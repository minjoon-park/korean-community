## 7. Visualization of circuits, states, and measurements: Practice Questions 

**1) What is the output of the following code?**

```
from qiskit import QuantumCircuit

qc = QuantumCircuit(3, 3)
qc.h(0)
qc.cx(0, 1)
qc.cx(0, 2)
qc.measure(range(3), range(3))
qc.draw(output="mpl", reverse_bits=True)
```

A) The circuit diagram is drawn with qubit 0 at the top and qubit 2 at the bottom

B) The circuit diagram is drawn with qubit 2 at the top and qubit 0 at the bottom

C) The circuit diagram is drawn with classical bits reversed but qubits unchanged

D) A ValueError is raised because reverse_bits is incompatible with measurement operations

***A:***
<Details>
B) By default Qiskit uses little-endian notation (qubit 0 at top). Setting reverse_bits=True reverses the display order so qubit 2 appears at the top.
<br/>
</Details>

**7) Consider this code:**
```
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere

qc = QuantumCircuit(2)
qc.h(0)
qc.z(0)
qc.x(1)
qc.cx(0, 1)
state = Statevector(qc)
plot_state_qsphere(state, show_state_phases=True, use_degrees=True)
```
**How many dots appear on the Q-sphere, and what are their locations?**

A) 2 dots (|01⟩ and |10⟩) on the equator, separated by 180°

B) 4 dots (all basis states) on the equator, separated by 90°

C) 2 dots (|00⟩ and |11⟩) on the north and south poles

D) 1 dot (|01⟩) at the north pole

***A:***
<Details>
A) The state (|01⟩ − |10⟩)/√2 has two non-zero amplitudes: |01⟩ with phase 0° and |10⟩ with phase 180° (due to the minus sign). Two dots appear on the equator of the Q-sphere with equal size and 180° separation.
<br/>
</Details>

**8) Given the following code, what does the Hinton diagram display?**
```
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_hinton

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
state = Statevector(qc)
plot_state_hinton(state)
```

A) A grid of squares whose sizes represent the magnitude of density matrix elements; white squares are positive, black squares are negative

B) A 3D bar chart of the real part of the statevector

C) A heatmap of qubit-qubit correlations

D) A Bloch sphere with a Hinton-style overlay

***A:***
<Details>
A) The Hinton diagram shows the density matrix as a grid of squares. Square size indicates magnitude and color (white/black) indicates sign (positive/negative).
<br/>
</Details>
