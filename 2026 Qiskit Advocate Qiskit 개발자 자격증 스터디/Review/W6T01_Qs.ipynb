## 1. Quantum operators and gate foundations: Practice Questions 

**1) Pauli Boolean Array Initialization: Given the code below, what is the output?**

```
from qiskit.quantum_info import Pauli

z = [True, True, False, False]
x = [True, False, True, False]
p = Pauli((z, x))
print(p.to_label())
```

A) YZXI

B) IXZY

C) XZYΙ

D) IХYZ


***A:***
<Details>
B) (string representation uses little endian.)

**2) Pauli from QuantumCircuit: Given the code below, what is the output?**

```
from qiskit import QuantumCircuit
from qiskit.quantum_info import Pauli

qc = QuantumCircuit(3)
qc.z(0)
qc.x(1)
qc.y(2)
p = Pauli(qc)
print(p.to_label())
```

A) ZXY

B) YXZ

C) ZYX

D) XYZ

***A:***
<Details>
B)

**3) Pauli Composition: Given the code below, what is the output?**

```
from qiskit.quantum_info import Pauli

result = Pauli('X').compose(Pauli('Y'))
print(result.to_label())
```

A) iZ

B) -iZ

C) Z

D) Y

***A:***
<Details>
B) Apply X, and then Y, i.e., YX = -iZ
<br/>
</Details>

**4) Pauli Delete Operation: Given the code below, what is the output?**

```
from qiskit.quantum_info import Pauli

p = Pauli("XYZI")
result = p.delete([0, 2])
print(result.to_label())
```

A) XZ

B) YI

C) XY

D) YX

***A:***
<Details>
A) Don't get fooled by little-endian-ness.

Extension: list multiple circuits, and ask "which one is different?"
<br/>
</Details>
