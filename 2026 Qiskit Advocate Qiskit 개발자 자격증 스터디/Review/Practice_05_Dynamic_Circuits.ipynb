**2) 다음 회로를 만드는 코드로 올바른 것은?**

![](./images/dynamic_circuit_02.png)

**A)**
```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister

qr = QuantumRegister(4, 'q')
cr0 = ClassicalRegister(3, 'c0')
cr1 = ClassicalRegister(1, 'c1')
circuit = QuantumCircuit(qr, cr0, cr1)

circuit.h(qr[0:3])
circuit.measure(qr[0:3], cr0)
with circuit.if_test(cr0[0]^cr0[1]^cr0[2]):
    circuit.x(qr[3])
circuit.measure(qr[3], cr1)
```

**B)**
```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister

qr = QuantumRegister(4, 'q')
cr0 = ClassicalRegister(3, 'c0')
cr1 = ClassicalRegister(1, 'c1')
circuit = QuantumCircuit(qr, cr0, cr1)

circuit.h(qr[0:3])
circuit.measure(qr[0:3], cr0)
if cr0[0]^cr0[1]^cr0[2] == 1:
    circuit.x(qr[3])
circuit.measure(qr[3], cr1)
```

**C)**
```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.circuit.classical import expr

qr = QuantumRegister(4, 'q')
cr0 = ClassicalRegister(3, 'c0')
cr1 = ClassicalRegister(1, 'c1')
circuit = QuantumCircuit(qr, cr0, cr1)

circuit.h(qr[0:3])
circuit.measure(qr[0:3], cr0)
with circuit.if_test(expr.bit_xor(expr.bit_xor(cr0[0], cr0[1]), cr0[2])):
    circuit.x(qr[3])
circuit.measure(qr[3], cr1)
```

**D)**
```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister

qr = QuantumRegister(4, 'q')
cr0 = ClassicalRegister(3, 'c0')
cr1 = ClassicalRegister(1, 'c1')
circuit = QuantumCircuit(qr, cr0, cr1)

circuit.h(qr[0:3])
circuit.measure(qr[0:3], cr0)
with circuit.if_test(cr0[0] xor cr0[1] xor cr0[2]):
    circuit.x(qr[3])
circuit.measure(qr[3], cr1)
```


***Answer:***
<Details>
C) Dynamic circuit에서 logical operation은 expr 함수를 통해 구현됩니다.
